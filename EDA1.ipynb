```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
columns=['age','workclass','fnlwgt',"education",'education-num','marital-status','occupation','relationship','race','sex','capital-gain','capital-loss','hours-per-week','native-country','>50K']
df = pd.read_csv(r"C:\Users\joshi\Downloads\adult.data",names=columns)
```

```python
df.head()
```

```python
#copying data_set
df_copy = df.copy()
df_copy.head()
```

```python
df_copy.info()
```

```python
# checking null data for each column in data set

print('Null data in the data set')
for i in columns:
    print(i,'----------------------',df_copy[i].isnull().sum(),'\n')

```

```python
# checking duplicate data
df_copy.duplicated().sum()
```

```python
df_copy[df_copy.duplicated()==True]
```

```python
# remoing duplicate data
df_copy = df_copy.drop_duplicates()
df_copy.duplicated().sum()
```

```python
# segregating data into categorical data and numerical data

numerical = []
categorical = []
for i in columns:
    if df_copy[i].dtypes =='int64':
        numerical.append(i)
    else:
        categorical.append(i)
```

```python
# each categories with high capital gain

age_occupation_edu_gain = df_copy.groupby(['age','workclass','education','marital-status','race','occupation','sex']).agg({'capital-gain':'sum'}).sort_values(by='capital-gain',ascending=False)
```
